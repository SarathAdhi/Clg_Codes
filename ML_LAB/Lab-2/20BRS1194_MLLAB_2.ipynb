```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
tst = pd.read_csv("wheat_test.csv")
tst.head()
```

```python
train = pd.read_csv("wheat_train.csv")
train.head()
```

```python
T = train.drop(['ID'],axis = 1)
T.head()
```

```python
t = tst.drop(['ID'],axis = 1)
t.head()
```

```python
X = T.drop('Type',axis=1)
y = T['Type']
```

```python
logistic = LogisticRegression()
logistic.fit(X,y)#fit means traning in machine learning,
#Fit function adjusts weights according to data values so that better accuracy can be achieved. 
#After training, the model can be used for predictions, using .predict() method call. 
```

```python
# output of an algorithm after it has been trained on a training dataset
x = t
predictions = logistic.predict(x)#what's happening here?
predictions
```

```python
logistic.score(x, predictions)#score of test data
```

```python
logistic.score(X, y)#score of training data
```
